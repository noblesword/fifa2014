```python
#importing libraries
import pandas as pd
import numpy as np
# read a csv from fifa_allcountries
fifa=pd.read_csv('FIFA2014 - all players.csv')
fifa.head()
```

```python
fifa.info()
```

```python
# Create Players.csv
players = fifa[['Player id', 'Player', 'Position', 'Number', 'D.O.B', 'Age', 'Height (cm)', 'Country', 'Caps', 'International goals']]
players.to_csv('Players.csv', index=False)
```

```python
# Create Clubs.csv
clubs = fifa[['Club', 'Club (country)']].drop_duplicates().rename(columns={'Club': 'Club_Name', 'Club (country)': 'Club_Country'})
clubs.to_csv('Clubs.csv', index=False)  
```

```python
# Create Plays_For.csv
plays_for = fifa[['Player id', 'Club', 'Plays in home country?']].rename(columns={'Club': 'Club_Name'})
plays_for.to_csv('Plays_For.csv', index=False)
```

```python
# Create Teammates.csv (this requires a bit more data processing)
from itertools import combinations

# Assume we process only for one club at a time to avoid memory issues
teammates = pd.DataFrame(columns=['Player ID 1', 'Player ID 2', 'Club_Name'])
for club in clubs['Club_Name'].unique():
    players_in_club = fifa[fifa['Club'] == club]['Player id']
    club_teammates = pd.DataFrame(list(combinations(players_in_club, 2)), columns=['Player ID 1', 'Player ID 2'])
    club_teammates['Club_Name'] = club
    teammates = pd.concat([teammates, club_teammates], ignore_index=True)

teammates.to_csv('Teammates.csv', index=False)
```
